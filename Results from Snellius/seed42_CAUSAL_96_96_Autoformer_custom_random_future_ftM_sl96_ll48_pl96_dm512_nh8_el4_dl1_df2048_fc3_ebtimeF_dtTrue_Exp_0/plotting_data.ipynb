```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
true_data = np.load('true.npy', allow_pickle=True)
pred_data = np.load('pred.npy', allow_pickle=True)

# Calculate the mean across the first dimension (if it makes sense for your data)
true_mean = np.mean(true_data, axis=0)
pred_mean = np.mean(pred_data, axis=0)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(true_mean, label='True Mean', marker='o')
plt.plot(pred_mean, label='Pred Mean', marker='x')
plt.title('Comparison of True and Predicted Data')
plt.xlabel('Sample Index')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
